```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/superstore_cleaned.csv', encoding='latin-1')
print(df.shape)
print(df.columns.tolist())
```

```python
conn = sqlite3.connect('../data/superstore.db')
df.to_sql('orders', conn, if_exists='replace', index=False)
print("done,", len(df), "rows loaded")
```

```python
q1 = pd.read_sql_query("SELECT * FROM orders LIMIT 5", conn)
q1
```

```python
q2 = pd.read_sql_query("""
    SELECT Category, 
           ROUND(SUM(Sales), 2) as total_sales,
           COUNT(*) as num_orders
    FROM orders
    GROUP BY Category
    ORDER BY total_sales DESC
""", conn)

q2
```

```python
plt.figure(figsize=(6,4))
plt.bar(q2['Category'], q2['total_sales'], color=['#4C9BE8','#E8834C','#4CB87E'])
plt.title('Total Sales by Category')
plt.ylabel('Sales ($)')
plt.tight_layout()
plt.savefig('../outputs/sql_category_sales.png')
plt.show()
```

```python
q3 = pd.read_sql_query("""
    SELECT 
        Category,
        Month,
        Year,
        SUM(Quantity) as monthly_units,
        AVG(SUM(Quantity)) OVER (
            PARTITION BY Category 
            ORDER BY Year, Month
            ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
        ) as rolling_avg
    FROM orders
    GROUP BY Category, Year, Month
    ORDER BY Category, Year, Month
""", conn)

q3.head(10)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=False)
cats = ['Furniture', 'Technology', 'Office Supplies']
colors = ['#E8834C', '#4C9BE8', '#4CB87E']

for ax, cat, col in zip(axes, cats, colors):
    sub = q3[q3['Category'] == cat]
    ax.bar(range(len(sub)), sub['monthly_units'], color=col, alpha=0.4, label='monthly')
    ax.plot(range(len(sub)), sub['rolling_avg'], color=col, linewidth=2, label='3m avg')
    ax.set_title(cat)
    ax.set_xlabel('Month')
    ax.set_ylabel('Units')
    ax.legend(fontsize=8)

plt.suptitle('Monthly Demand with 3-Month Rolling Average', fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/sql_demand_trend.png')
plt.show()
```

```python
q4 = pd.read_sql_query("""
    WITH monthly AS (
        SELECT [Sub-Category],
               SUM(Sales) as monthly_sales
        FROM orders
        GROUP BY [Sub-Category], Year, Month
    ),
    summary AS (
        SELECT [Sub-Category],
               SUM(monthly_sales)  as total_sales,
               AVG(monthly_sales)  as avg_monthly_sales
        FROM monthly
        GROUP BY [Sub-Category]
    )
    SELECT [Sub-Category],
           ROUND(total_sales, 0) as total_sales,
           ROUND(avg_monthly_sales, 0) as avg_monthly,
           ROUND(total_sales / avg_monthly_sales, 1) as turnover_ratio,
           CASE 
               WHEN total_sales / avg_monthly_sales >= 8 THEN 'Fast'
               WHEN total_sales / avg_monthly_sales >= 4 THEN 'Medium'
               ELSE 'Slow'
           END as speed
    FROM summary
    ORDER BY turnover_ratio DESC
""", conn)

q4
```

```python
plt.figure(figsize=(10, 5))
colors = ['#4CB87E' if s == 'Fast' else '#E8C24C' if s == 'Medium' else '#E8554C' 
          for s in q4['speed']]
plt.barh(q4['Sub-Category'], q4['turnover_ratio'], color=colors)
plt.xlabel('Turnover Ratio')
plt.title('Inventory Turnover by Sub-Category')
plt.tight_layout()
plt.savefig('../outputs/sql_turnover.png')
plt.show()
```

```python
q5 = pd.read_sql_query("""
    SELECT [Ship Mode],
           Region,
           COUNT(*) as total_orders,
           ROUND(AVG(Delivery_Days), 1) as avg_days,
           ROUND(100.0 * SUM(CASE WHEN Delivery_Days <= 3 THEN 1 ELSE 0 END) 
                 / COUNT(*), 1) as pct_on_time
    FROM orders
    GROUP BY [Ship Mode], Region
    ORDER BY avg_days DESC
""", conn)

q5
```

```python
pivot = q5.pivot(index='Region', columns='Ship Mode', values='avg_days')

pivot.plot(kind='bar', figsize=(9, 5), colormap='Set2', edgecolor='none')
plt.title('Avg Delivery Days by Region and Ship Mode')
plt.ylabel('Avg Days')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.legend(title='Ship Mode', fontsize=8)
plt.axhline(y=3, color='red', linestyle='--', linewidth=1, label='3-day target')
plt.tight_layout()
plt.savefig('../outputs/sql_lead_time.png')
plt.show()
```

```python
q6 = pd.read_sql_query("""
    SELECT Category,
           [Sub-Category],
           ROUND(SUM(Profit), 2) as total_profit,
           ROUND(AVG(Discount) * 100, 1) as avg_discount_pct,
           ROUND(100.0 * SUM(Profit) / SUM(Sales), 1) as margin_pct,
           RANK() OVER (
               PARTITION BY Category 
               ORDER BY SUM(Profit) DESC
           ) as rank_in_category
    FROM orders
    GROUP BY Category, [Sub-Category]
    ORDER BY Category, rank_in_category
""", conn)

q6
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#4CB87E' if x > 0 else '#E8554C' for x in q6['total_profit']]
ax.barh(q6['Sub-Category'], q6['margin_pct'], color=colors)
ax.axvline(x=0, color='black', linewidth=0.8)
ax.set_xlabel('Profit Margin %')
ax.set_title('Profit Margin % by Sub-Category')
plt.tight_layout()
plt.savefig('../outputs/sql_profit_margin.png')
plt.show()
```

```python
q7 = pd.read_sql_query("""
    SELECT 
        COUNT(*) as total_rows,
        SUM(CASE WHEN [Order ID] IS NULL THEN 1 ELSE 0 END) as null_order_id,
        SUM(CASE WHEN Sales < 0 THEN 1 ELSE 0 END) as negative_sales,
        SUM(CASE WHEN Quantity <= 0 THEN 1 ELSE 0 END) as bad_quantity,
        SUM(CASE WHEN Discount < 0 OR Discount > 1 THEN 1 ELSE 0 END) as invalid_discount
    FROM orders
""", conn)

q7
```

```python
conn.close()
print("done")
```
