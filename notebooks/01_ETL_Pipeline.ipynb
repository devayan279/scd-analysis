```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin-1')

print("Data loaded successfully!")
print("Shape:", df.shape)
```

```python

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.strftime('%b')  # Jan, Feb...

print("Date columns fixed!")
print(df[['Order Date', 'Year', 'Month', 'Month_Name']].head())
```

```python
df['Delivery_Days'] = (df['Ship Date'] - df['Order Date']).dt.days

print("Average delivery days:", df['Delivery_Days'].mean().round(2))
print("Max delivery days:", df['Delivery_Days'].max())
print("Min delivery days:", df['Delivery_Days'].min())
```

```python
regional = df.groupby('Region').agg(
    Total_Sales = ('Sales', 'sum'),
    Total_Orders = ('Order ID', 'count'),
    Avg_Profit = ('Profit', 'mean')
).round(2)

print(regional)
```

```python

df.to_csv('../data/superstore_cleaned.csv', index=False)
print("Cleaned data saved successfully!")
print("New shape with added columns:", df.shape)
```
